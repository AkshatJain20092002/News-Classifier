```python
file = "E:/NLP Personal Practise/Assignments/News classification/All.csv"
df = pd.read_csv(file)
df
```

```python
df = df.drop(columns = ['title','Unnamed: 0'], axis = 0)
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
len(df)
```

```python
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
def tokenizewords(text):
    return ' '.join(word_tokenize(text))

df['text'] = df['text'].apply(tokenizewords)
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemwords(text):
    return ' '.join([ps.stem(word) for word in text.split()])

df['text'] = df['text'].apply(stemwords)
df.head()
```

```python
from nltk.corpus import stopwords
sw_list = stopwords.words('english')

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word.lower() not in sw_list])

df['text'] = df['text'].apply(remove_stopwords)
df.head()
```

```python
#Save it takes heck lot of time
```

```python
df.to_csv('E:/NLP Personal Practise/Assignments/News classification/processed_data.csv', index=False)
```

```python
X = df['text']
Y = df['label']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 0)

print(X)
print(Y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df = 0.7)

tfidf_train = tfidf.fit_transform(X_train)
tfidf_test = tfidf.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(solver='liblinear', penalty = 'l1')
model1.fit(tfidf_train, Y_train)
```

```python
Y_pred = model1.predict(tfidf_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm= confusion_matrix(Y_test, Y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
accuracy
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
model2 = PassiveAggressiveClassifier(max_iter = 50)
model2.fit(tfidf_train, Y_train)
```

```python
Y_pred = model2.predict(tfidf_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm= confusion_matrix(Y_test, Y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
accuracy
```
